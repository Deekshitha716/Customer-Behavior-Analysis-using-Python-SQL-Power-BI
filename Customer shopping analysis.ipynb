```python
import pandas as pd
```

```python
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] =  df.groupby('Category')['Review Rating'].transform(lambda x : x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ', "_")
df = df.rename(columns={'purchase_amount_(usd)' :'purchase_amount'})
```

```python
df.columns
```

<h1>Feature Engineering</h1>

```python
df['age_groups'] = pd.qcut(df['age'],q=4, labels = ['Young', 'Adult', 'Middle-aged', 'Senior'])
```

```python
df[['age', 'age_groups']]
```

```python
frequency_dict = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Annually' : 365,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Every 3 Months' : 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_dict)
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']]
```

```python
df[['discount_applied', 'promo_code_used']]
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
df = df.drop('promo_code_used', axis = 'columns')
```
